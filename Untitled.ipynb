```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.linspace(0,100,1001)
w
```

```python
def complexn(w, wp, w0, g):
    return(wp/(w0**2-1j*g*w-w**2))
```

```python
ntilde = complexn(w, 10, 20, 1)
```

```python
ntilde
```

```python
fig0, ax0 = plt.subplots()
ax0.plot(w, ntilde.real)
ax0.plot(w, ntilde.imag)
```
