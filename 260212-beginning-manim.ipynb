```python
%load_ext manim

%%manim -qm OscillatingChargeRadiation
from manim import *
import numpy as np

class OscillatingChargeRadiation(Scene):
    def construct(self):
        # parameters (feel free to tweak)
        A = 1.5          # oscillation amplitude (scene units)
        omega = 2.2      # angular freq
        c = 2.6          # wave speed (scene units / sec)
        r_max = 10.0
        dr = 0.7

        t = ValueTracker(0.0)

        # moving charge
        charge = always_redraw(lambda: Dot(
            point=np.array([0.0, A*np.sin(omega*t.get_value()), 0.0]),
            radius=0.12
        ))

        # a faint axis line for orientation
        axis = Line(LEFT*7, RIGHT*7).set_stroke(width=1, opacity=0.25)

        # "radiation": expanding circles with retarded-time modulation
        def wave_circles():
            tt = t.get_value()
            circles = VGroup()
            for r in np.arange(0.5, r_max, dr):
                phase = omega*(tt - r/c)          # retarded time
                amp = 0.5*(1 + np.sin(phase))     # in [0,1]
                op = 0.05 + 0.35*amp              # opacity modulation
                w  = 1.0  + 2.0*amp               # stroke width modulation
                circles.add(Circle(radius=r).set_stroke(width=w, opacity=op))
            return circles

        waves = always_redraw(wave_circles)

        self.add(axis, waves, charge)
        self.play(t.animate.set_value(6.0), run_time=6, rate_func=linear)
```
