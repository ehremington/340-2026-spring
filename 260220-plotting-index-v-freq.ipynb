```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.linspace(1,40,1001)
```

```python
def complexn(w, wp, w0, g):
    return(np.sqrt(1+wp**2/(w0**2-1j*g*w-w**2)))
```

```python
ntilde = complexn(w, 10, 0, 1)
```

```python
ntilde
```

```python
fig0, ax0 = plt.subplots()
ax0.plot(w-10, ntilde.real)
ax0.plot(w-10, ntilde.imag)
```
