```python
%matplotlib widget

from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rectifiedSquareTarget(x):
    # 2π-periodic rectangular wave centered on x = 0
    xwrap = (x + np.pi) % (2*np.pi) - np.pi
    return np.where(np.abs(xwrap) <= np.pi/2, 1.0, 0.0)

def rectifiedSquarePartial(x, N):
    # N = number of nonzero Fourier terms
    y = 0.5*np.ones_like(x)
    
    for m in range(N):
        n = 2*m + 1   # only odd harmonics survive
        y += (2/np.pi) * ((-1)**m / n) * np.cos(n*x)
    
    return y

def fourierRectDemo(wavefunc):
    fig0 = plt.figure(figsize=(8,5))
    ax0 = fig0.add_subplot(111)
    
    x = np.linspace(-2*np.pi, 2*np.pi, 2000)
    
    line01, = ax0.plot(x, rectifiedSquareTarget(x), color='gray', linewidth=2, label='target wave')
    line02, = ax0.plot([], [], linewidth=2, label='Fourier sum')
    
    ax0.set_xlim((-2*np.pi, 2*np.pi))
    ax0.set_ylim((-0.25, 1.25))
    ax0.axhline(0, color='k')
    ax0.axvline(0, color='k')
    ax0.set_xlabel('x')
    ax0.set_ylabel('f(x)')
    ax0.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
    ax0.set_xticklabels([r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$'])
    ax0.legend()
    
    def plot0(N=0):
        wave2 = wavefunc(x, N)
        line02.set_data(x, wave2)
        ax0.set_title(f'{N} Fourier terms')
        plt.draw()
    
    interact(plot0, N=(0, 100, 1))

fourierRectDemo(rectifiedSquarePartial)
```
